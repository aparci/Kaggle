Since all of three passengers from above have same ticket number we can asume that their fare price is calculated for group, so we need to divide fare on three of them. The same thing happened with passengers that have fare above 200. According to that I will make function that counts how many people have the same ticket and divides fare according to number of people.

```python
group_fare = train[train['Fare'] > 60].groupby(by = 'Ticket').size().to_frame(name='counts')
group_fare # this would be frame used for fare correction
```

```python
group_fare.shape
```

```python
idx = train.index[train['Ticket'] == 'PC 17755'][0]
idx #trying to find index of person with certain index
```

```python
group_fare['Ticket'] = group_fare.index
group_fare.head()
```

```python
group_fare.reset_index(drop = True)
group_fare.head()
```

```python
group_fare.iloc[2]['Ticket']
```

```python
def fare_correction (df):
    rows = df.shape[0]
    for i in range(0, rows):
        if df.iloc[i]['counts'] > 1:
            idx = train.index[train['Ticket'] == df.iloc[i]['Ticket']][0]
            new_fare = train.iloc[idx]['Fare'] / df.iloc[i]['counts']
            train.loc[train['Ticket'] == df.iloc[i]['Ticket'], 'Fare'] = new_fare
    
```

```python
# Function to generate submission file to get test score
def submission(preds):
    test['Survived'] = preds
    predictions = test[['PassengerId', 'Survived']]
    predictions.to_csv('submission.csv', index=False)
```

## Modeling<a class="anchor" id="third-bullet"></a>

### Logistic Regression

```python
# Classification model
LogReg = LogisticRegression()
```

```python
seed = 101

# Parameters to tune
params = [{'penalty': ['l1', 'l2', 'elasticnet', 'none'],
           'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits = 5, n_repeats = 10, random_state = seed)
lr_clf = GridSearchCV(LogReg, params, cv = cv, n_jobs = -1)
lr_clf.fit(X_train, y_train)
```

```python
# Best parameters
lr_clf.best_params_
```

```python
# Train score
lr_clf.best_score_
```

```python
# Test score
y_preds = lr_clf.predict(X_test)
submission(y_preds)
```

### KNN

```python
# Classification model
knn = KNeighborsClassifier()

# Parameters to tune
params = [{'n_neighbors': range(1, 21),
           'p': [1, 2]}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=seed)
knn_clf = GridSearchCV(knn, params, cv=cv, n_jobs=-1)
knn_clf.fit(X_train, y_train)
```

```python
# Best parameters
knn_clf.best_params_
```

```python
# Train score
knn_clf.best_score_
```

```python
# Test score
y_preds = knn_clf.predict(X_test)
submission(y_preds)
```

### SVM

```python
# Classification model
svm = SVC(max_iter = 10000)

# Parameters to tune
params = [{'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
           'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits = 5, n_repeats = 10, random_state = seed)
svm_clf = GridSearchCV(svm, params, cv = cv, n_jobs = -1)
svm_clf.fit(X_train, y_train)
```

```python
# Best parameters
svm_clf.best_params_
```

```python
# Train score
svm_clf.best_score_
```

```python
# Test score
y_preds = svm_clf.predict(X_test)
submission(y_preds)
```

### Decision Tree

```python
# Classification model
dt = DecisionTreeClassifier(random_state=seed)

# Parameters to tune
params = [{'max_depth': [5, 7, 10, None],
           'min_samples_split': [2, 5, 10],
           'max_features': ['sqrt', 5, 7, 10]}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits = 5, n_repeats = 10, random_state=seed)
dt_clf = GridSearchCV(dt, params, cv = cv, n_jobs = -1)
dt_clf.fit(X_train, y_train)
```

```python
# Best parameters
dt_clf.best_params_
```

```python
# Train score
dt_clf.best_score_
```

```python
# Test score
y_preds = dt_clf.predict(X_test)
submission(y_preds)
```

### Random Forest

```python
# Note: This cell will take a while to run depending on the available processing power

# Classification model
rf = RandomForestClassifier(random_state=seed)

# Parameters to tune
params = [{'n_estimators': range(50, 550, 50),
           'max_depth': [5, 7, 10, None],
           'min_samples_split': [2, 5, 10],
           'max_features': ['sqrt', 5, 7, 10]}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=seed)
rf_clf = GridSearchCV(rf, params, cv=cv, n_jobs=-1)
rf_clf.fit(X_train, y_train)
```

```python
# Best parameters
rf_clf.best_params_
```

```python
# Train score
rf_clf.best_score_
```

```python
# Test score
y_preds = rf_clf.predict(X_test)
submission(y_preds)
```

### XGB Classification

```python
# Note: This cell will take a while to run depending on the available processing power

# Classification model
xgboost = xgb.XGBClassifier(random_state=seed)

# Parameters to tune
params = [{'max_depth': [3, 5, 10],
           'learning_rate': [0.01, 0.02, 0.03, 0.05, 0.07, 0.1],
           'n_estimators': range(100, 1100, 100)}]

# Hyperparameter tuning using GridSearchCV
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=seed)
xgb_clf = GridSearchCV(xgboost, params, cv=cv, n_jobs=-1)
xgb_clf.fit(X_train, y_train)
```

```python
# Best parameters
xgb_clf.best_params_
```

```python
# Train score
xgb_clf.best_score_
```

```python
# Test score
y_preds = xgb_clf.predict(X_test)
submission(y_preds)
```

### Model Stacking

```python
# Models that we will input to stacking classifier
base_estimators = list()
base_estimators.append(('lda', lda))
base_estimators.append(('knn', knn_clf.best_estimator_))
base_estimators.append(('svm', svm_clf.best_estimator_))
base_estimators.append(('dt', dt_clf.best_estimator_))
base_estimators.append(('rf', rf_clf.best_estimator_))

# Stacking classifier
stacking_clf = StackingClassifier(estimators=base_estimators, final_estimator=LogisticRegression(), cv=5, n_jobs=-1)
stacking_clf.fit(X_train, y_train)
```

```python
# Test score
y_preds = stacking_clf.predict(X_test)
submission(y_preds)
```
